Fill Numeric Columns Using Median

```python
data['likes'] =  data['likes'].fillna(data['likes'].median())
data['comments']  =  data['comments'].fillna(data['comments'].median())
data['shares'] = data['shares'].fillna(data['shares'].median())
data['age'] = data['age'].fillna(data['age'].median())
```

Fill Categorical Data Using Mode

```python
data['gender'] = data['gender'].fillna(data['gender'].mode()[0])
data['sentiment'] = data['sentiment'].fillna(data['sentiment'].mode()[0])
```

**Data Visualization**

```python
import matplotlib.pyplot as plt
import seaborn as  sns
import  plotly.express as px
```

Scatter Plot

```python
plt.figure(figsize=(8,5))
plt.scatter(data['likes'], data['impression_count'])

plt.xlabel("likes")
plt.ylabel("Impressions")
plt.title("Likes vs Impressions")

plt.show()

```

This scatter plot shows the relationship between the number of likes and impressions. It can be observed that posts with higher impressions generally receive more likes. This indicates a positive relationship between visibility and user engagement. However, some posts have high impressions but relatively lower likes, suggesting that visibility alone does not always guarantee strong engagement.

**`Line Chart`**
Daily Engagement Trend

```python
daily = data.groupby('posted_at')['engagement_rate'].mean()
```

```python
plt.figure(figsize=(10,5))
plt.plot(daily)
plt.title("Daily Engagement Trend")
plt.xlabel("Date")
plt.ylabel("Average Engagement Rate")
plt.show()
```

The line chart shows how the average engagement rate changes over time. From the visualization, engagement levels fluctuate across different days rather than remaining constant. Some days show higher engagement, which may indicate that users are more active during certain periods. Monitoring these patterns can help determine the best days to post content.




**Bar Chart**

Posts by Category

```python
data['post_category'].value_counts().plot(kind='bar')

plt.title("Number of post by Category")
plt.xlabel("Category")
plt.ylabel("Count")

plt.show()
```

This bar chart displays the distribution of posts across different content categories. It shows which type of content is posted more frequently. Categories with higher counts represent the main focus of the platform's content strategy, while categories with fewer posts may indicate areas that are less utilized.

**Pie Chart**

Gender Distribution

```python
data['gender'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title("Gender Distribution")
plt.show()
```

The pie chart represents the percentage distribution of users based on gender. It helps in understanding the audience demographic. If one gender has a larger share, it suggests that the platform attracts more engagement from that group, which could influence future content targeting.

**Histogram (Matplotlib)**
Age Distribution

```python
plt.figure(figsize=(8,5))
plt.hist(data['age'], bins=20)
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")

plt.show()
```

The histogram shows how users are distributed across different age groups. The visualization indicates that most users fall within a certain age range, forming the peak of the distribution. This helps identify the dominant age group interacting with the platform and can guide content creation strategies.

**Count Plot (Seaborn)**
Post Type Distribution

```python
sns.countplot(x='post_type', data=data)
plt.title("Post Type Distribution")
plt.show()
```

This count plot displays how frequently different types of posts are published. Some post types appear more often than others, indicating the preferred format used on the platform. Understanding which post types are commonly used can help analyze whether certain formats attract more engagement.

**Violin Plot (Seaborn)**

Followers vs Sentiment

```python
sns.violinplot(x='sentiment', y='follower_count', data=data)
plt.title("Follower Distribution by Sentiment")
plt.show()
```

The violin plot shows how follower counts vary across different sentiment categories. It provides insight into whether positive, neutral, or negative sentiment posts are associated with higher follower counts. Wider sections in the plot indicate where follower counts are more concentrated.

**Pair Plot (Seaborn)**

Numeric Feature Relationships

```python
sns.pairplot(data[['likes','comments','shares','engagement_rate']])
plt.show()
```

The pairplot visualizes relationships between likes, comments, shares, and engagement rate. It helps identify patterns and correlations among these engagement metrics. For example, posts with higher likes may also receive more comments and shares, suggesting that these engagement actions are often related.

**Interactive Plot (Plotly)**

```python
fig = px.scatter(data,
                 x='likes',
                 y='impression_count',
                 color='post_type',
                 size='engagement_rate',
                 title='Interactive Engagement Analysis')

fig.show()
```

The heatmap displays the correlation between numerical variables in the dataset. Strong positive correlations indicate that two variables tend to increase together, while weaker correlations show little relationship. This visualization helps identify which engagement metrics are closely connected and influence each other.

**Heatmap (Seaborn)**

Correlation Matrix

```python
plt.figure(figsize=(10,7))
corr = data.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("Correlation Matrix")
plt.show()
```

The heatmap shows the correlation between different numerical variables in the dataset. It can be observed that likes, comments, and shares have a positive correlation with the engagement rate, meaning posts with more interactions usually have higher engagement. Some variables show weak or low correlation, which indicates they do not strongly influence each other.
